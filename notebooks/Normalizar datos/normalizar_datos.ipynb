```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import os

# 1. Definir rutas (usamos el prefijo 'r' para leer correctamente las rutas en Windows)
ruta_entrada = r"..\..\data\raw\data_set_1\dataset_orinoco_multivariado_final copy.csv"
ruta_salida_dir = r"..\..\data\processed"
ruta_salida_archivo = os.path.join(ruta_salida_dir, "dataset_normalizado.csv")

# 2. Cargar datos
df = pd.read_csv(ruta_entrada)

# 3. Procesar la fecha
# Asegurarse de que la columna fecha esté en formato datetime
df['fecha'] = pd.to_datetime(df['fecha'])

# Extraer año, mes y semana
df['año'] = df['fecha'].dt.year
df['mes'] = df['fecha'].dt.month
df['semana'] = df['fecha'].dt.isocalendar().week.astype(int)

# Transformación cíclica para mes (ciclo de 12)
df['mes_sin'] = np.sin(2 * np.pi * df['mes'] / 12)
df['mes_cos'] = np.cos(2 * np.pi * df['mes'] / 12)

# Transformación cíclica para semana (ciclo de 52)
df['semana_sin'] = np.sin(2 * np.pi * df['semana'] / 52)
df['semana_cos'] = np.cos(2 * np.pi * df['semana'] / 52)

# Eliminar la fecha original y las componentes lineales (ya representadas en ciclo)
df = df.drop(columns=['fecha', 'mes', 'semana'])

# 4. Normalizar el resto de los datos con Min-Max Scaler (incluyendo el 'año')
# Seleccionamos todas las columnas excepto las cíclicas (estas ya están acotadas entre -1 y 1)
columnas_ciclicas = ['mes_sin', 'mes_cos', 'semana_sin', 'semana_cos']
columnas_a_escalar = [col for col in df.columns if col not in columnas_ciclicas]

scaler = MinMaxScaler()
df[columnas_a_escalar] = scaler.fit_transform(df[columnas_a_escalar])

# 5. Guardar el dataset resultante
# Asegurar de que la carpeta de destino exista
os.makedirs(ruta_salida_dir, exist_ok=True)

# Guardar
df.to_csv(ruta_salida_archivo, index=False)

print(f"Procesamiento exitoso. Archivo guardado en:\n{ruta_salida_archivo}")
```
